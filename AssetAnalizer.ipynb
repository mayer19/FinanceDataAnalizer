```python
#imports
import pandas_datareader.data as web
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def get_asset_data(ticker, start_date, end_date):
    """
    Function to download asset/stock from Yahoo Finance
    It needs a ticker str() and star and end date of the dataset.
    """
    #Override the default method in pandas_datareader to use yfinance
    yf.pdr_override()
    #Download data from Yahoo Finance
    data = web.get_data_yahoo(ticker, start=start_date, end=end_date)
    return data
```

```python
#Calculate data percentual change
data_percent_change = data['Adj Close'].pct_change() * 100 #convert to percent
data_percent_change.iloc[-1]
```

```python
#Corretation Matrix
data_percent_change.corr()
```

```python
# Plot correlation matrix using Seaborn
plt.figure(figsize=(10, 20))
sns.heatmap(data_percent_change.corr(), annot=True, cmap='coolwarm', center=0, vmin=-1, vmax=1)
plt.title('Correlation Matrix of Stock Returns')
plt.show()
```
